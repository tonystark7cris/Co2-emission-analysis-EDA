# In this notebook, I will show you evolution of CO2 emission by countries and will see which are the major countries responsible.

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

GreenHouse Gases emission from 1750 to 2019

```python
df = pd.read_csv('Book1.csv')
df.head()
```

```python
da = df.T
da.head()
```

```python
da = da.reset_index() # set a default index
da.head()
```

```python
da.columns = da.iloc[0]
da.head()
```

```python
da = da.rename(columns = {'country_name':'Year'})
da.head()
```

```python
da.drop(0, inplace =True)
da.head()
```

```python
da.info()
```

```python
da.index = pd.to_datetime(da['Year'])
da.drop('Year', axis=1, inplace=True)
da
```

```python
type(da.index)
```

# Which country has the highest greenhouse gases emission nowadays?

```python
sns.set(rc={'figure.figsize':(12, 5)})
axes = da[['United States','EU28','China','India','Russia','Australia']].loc['2000-1-1':].plot(marker='.')
axes.set_ylabel('GHG tonnes')
axes.set_title("The increasing of the emissions that happened in China is the biggest in the last 30 years. The GHG emission for China become higher than Russia after 2004");
```

# How much the proportion of the GHG emission has increased in the last 30 years?

```python
da90 = da.loc['1990-1-1']
labels = 'China','Russia','Europe','United States','India'
sizes = [da90['China'], da90['Russia'], da90['EU28'], da90['United States'],da90['India']]
explode = (0, 0, 0, 0.1,0)  
plt.style.use('dark_background')
fig1, ax = plt.subplots()
ax.pie(sizes, explode=explode, labels=labels, autopct='%1.f%%',shadow=True, startangle=120)
ax.axis('equal')  
ax.set_title('The proprtions of GHG emission in 1990');
```

```python
da90 = da.loc['2019-1-1']
labels = 'China','Russia','Europe','United States','India'
sizes = [da90['China'], da90['Russia'], da90['EU28'], da90['United States'],da90['India']]
explode = (0.1, 0, 0, 0,0)  
plt.style.use('dark_background')
fig1, ax = plt.subplots()
ax.pie(sizes, explode=explode, labels=labels, autopct='%1.f%%',shadow=True, startangle=120)
ax.axis('equal')  
ax.set_title('The proprtions of GHG emission in 2019');
```

```python
da[['United States','EU28','China','Russia','Japan','India']].loc['1990-1-1'].plot(kind="barh")
plt.title("The GHG emission in 2001")
plt.ylabel("Country")
plt.xlabel("GHG tonnes");
```

```python
da[['United States','EU28','China','Russia','Japan','India']].loc['2019-1-1'].plot(kind="barh")
plt.title("The GHG emission in 2019")
plt.ylabel("Country")
plt.xlabel("GHG tonnes");
```

# Conclusion
In the last 30 years, The GHG emission has increased for the industrial countries and China tend to have the highest proportion nowadays.
